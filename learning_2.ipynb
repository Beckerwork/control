  <h2>Die Vorhersagetechnik</h2>

  <img src="src/images/datapoints.jpg" style="float: right; margin-left: 16px; margin-bottom: 8px; max-width: 300px; border-radius: 6px;">

  <p>Jeder Punkt im Diagramm steht für ein Kind: Die x-Position zeigt die Körpergröße, die y-Position die gemessene Geschwindigkeit.</p>
  <p>Um Vorhersagen zu machen, brauchen wir eine Gerade, die möglichst gut durch diese Punktwolke läuft. Diese Gerade nennt man <strong>Regressionsgerade</strong>. Sie fasst den allgemeinen Zusammenhang zwischen Körpergröße und Geschwindigkeit zusammen.</p>
  <p>Mit dieser Geraden können wir dann für jede beliebige Körpergröße eine Geschwindigkeit vorhersagen – auch für Werte, die gar nicht in unseren gesammelten Daten vorkommen.</p>
  <p>Beispiel: Nehmen wir an, wir wollen die Geschwindigkeit für ein Kind mit 145 cm Körpergröße vorhersagen. Wir schauen, wo die Gerade bei x = 145 cm liegt, und lesen den entsprechenden y-Wert ab. Das ist unsere Vorhersage!</p>
  <p>Wichtig: Eine Regressionsgerade ist nur so gut wie die Daten, auf denen sie basiert. Und sie macht Durchschnittsvorhersagen. Dabei sind individuelle Abweichungen normal.</p>
