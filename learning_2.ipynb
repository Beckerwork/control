  <h2>Die Vorhersagetechnik</h2>

  <img src="src/images/datapoints.jpg" style="float: right; margin-left: 16px; margin-bottom: 8px; max-width: 300px; border-radius: 6px;">

  <p>Jeder Punkt im Diagramm steht für ein Kind. Dabei wurden zwei Dinge gemessen. Die Körpergröße und die maximale Laufgeschwindigkeit.</p>
  <p>Um mit diesen Datenpunkten Vorhersagen zu machen, wird eine Gerade durch diese Punkte gelegt. Die Gerade muss so gelegt werden, dass sie möglichst nah an allen Datenpunkten liegt. Mit dieser Geraden kann dann für jede beliebige Körpergröße eine Geschwindigkeit vorhergesagt werden. Auch wenn eine bestimmte Größe nicht im Datensatz vorkommt.</p>
  <p>Beispiel: Angenommen, es soll die Geschwindigkeit für ein Kind mit einer Körpergröße von 145 cm Körpergröße vorhergesagt werden. Bei einer Größe von 145 cm gibt es keinen Datenpunkt. Die Gerade läuft dann trotzdem durch den Bereich. Dann kann damit eine Vorhersage getroffen werden. So eine Gerade nennt man eine <strong>Regressionsgerade</strong>. Was Regression bedeutet, wird später erklärt.</p>
  <div style="clear: both;"></div>
</div>
